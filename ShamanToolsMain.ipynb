```python
# Base de datos de los productos
productos = [
    {"ID": 1, "Nombre": "Ladrillo 10x20x30", "PrecioCompra": 1200, "PrecioVenta": 1500, "CantidadDisponible": 1000, "TipoUnidad": "un"},
    {"ID": 2, "Nombre": "Cemento Gris x50kg", "PrecioCompra": 28000, "PrecioVenta": 30000, "CantidadDisponible": 200, "TipoUnidad": "saco"},
    {"ID": 3, "Nombre": "Varilla Corrugada #5 x6m", "PrecioCompra": 33000, "PrecioVenta": 35000, "CantidadDisponible": 150, "TipoUnidad": "un"},
    {"ID": 4, "Nombre": "Impermeabilizante 22kg", "PrecioCompra": 350000, "PrecioVenta": 360000, "CantidadDisponible": 50, "TipoUnidad": "un"},
    {"ID": 5, "Nombre": "Malla Electro Soldada 6x2.2m", "PrecioCompra": 42000, "PrecioVenta": 45000, "CantidadDisponible": 75, "TipoUnidad": "un"}
]

# Base de datos de los usuarios
usuarios = [
    {"ID": 1, "Usuario": "Ren Tao", "Contraseña": "1234_4321", "Rol": "admin"},
    {"ID": 2, "Usuario": "Ioh Asakura", "Contraseña": "abc_cba", "Rol": "vendedor"},
    {"ID": 3, "Usuario": "Chocolove McDonnell", "Contraseña": "11110000", "Rol": "vendedor"}
]

# Resumen de ventas
ventas = []

# Función para iniciar sesión
def iniciar_sesion():
    nombre_usuario = input("Ingrese su nombre de usuario: ")
    for usuario in usuarios:
        if usuario["Usuario"] == nombre_usuario:
            contraseña = input("Ingrese su contraseña: ")
            if usuario["Contraseña"] == contraseña:
                return usuario
            else:
                print("Contraseña incorrecta")
                return None
    print("Usuario no encontrado")
    return None

# Función para agregar o actualizar productos
def agregar_actualizar_producto():
    modificar_respuesta = input("desea modificar todo el producto, o solo la cantidad del total?\n1. modificar todo\n2. modificar solo la cantidad\nSeleccione una opción: ")

    if modificar_respuesta == "1":
        id_producto = int(input("Ingrese ID del producto: "))
        nombre = input("Ingrese nombre del producto: ")
        precio_compra = float(input("Ingrese precio de compra: "))
        precio_venta = float(input("Ingrese precio de venta: "))
        cantidad = int(input("Ingrese la cantidad que desea añadir: "))

        # Verificar si el producto ya existe
        for producto in productos:
            if producto["ID"] == id_producto:
                producto["Nombre"] = nombre
                producto["PrecioCompra"] = precio_compra
                producto["PrecioVenta"] = precio_venta
                producto["CantidadDisponible"] += cantidad  # Aumentar la cantidad
                print("Producto actualizado exitosamente.")
                return

    elif modificar_respuesta == "2":
        id_producto = int(input("Ingrese ID del producto a modificar: ")) # preguntar la ID del producto
        cantidad = int(input("Ingrese la cantidad que desea añadir: "))
        for producto in productos:  # iterar hazta encontrar el producto
            if producto["ID"] == id_producto:
                producto["CantidadDisponible"] += cantidad  # Aumentar la cantidad
                print("Producto actualizado exitosamente.")
                return
        print("Producto no encontrado.")
        return  # salimos de la


    # Si no existe, agregar el nuevo producto
    nuevo_producto = {
        "ID": id_producto,
        "Nombre": nombre,
        "PrecioCompra": precio_compra,
        "PrecioVenta": precio_venta,
        "CantidadDisponible": cantidad,
        "TipoUnidad": input("Ingrese tipo de unidad: ")
    }
    productos.append(nuevo_producto)
    print("Producto agregado exitosamente.")

# Función para consultar inventario por código o listar todos
def consultar_inventario():
    opcion = int(input("1. Buscar por código\n2. Listar todos los productos\nSeleccione una opción: "))
    if opcion == 1:
        id_producto = int(input("Ingrese ID del producto a buscar: "))
        for producto in productos:
            if producto["ID"] == id_producto:
                print(producto)
                return
        print("Producto no encontrado.")
    elif opcion == 2:
        for producto in productos:
            print(producto)
    else:
        print("Opción no válida.")

# Función para registrar venta
def registrar_venta(usuario):
    id_producto = int(input("Ingrese ID del producto: "))
    cantidad_vender = int(input("Ingrese cantidad a vender: "))
    for producto in productos:
        if producto["ID"] == id_producto:
            if producto["CantidadDisponible"] >= cantidad_vender:
                producto["CantidadDisponible"] -= cantidad_vender
                total_generado = producto["PrecioVenta"] * cantidad_vender
                ventas.append({"IDProducto": id_producto, "Cantidad": cantidad_vender, "Vendedor": usuario["Usuario"], "Total": total_generado})
                print("Venta registrada exitosamente.")
            else:
                print("Cantidad insuficiente en inventario.")
            return
    print("Producto no encontrado.")

# Función para registrar nuevos usuarios
def registrar_usuario():
    id_usuario = len(usuarios) + 1
    nombre_usuario = input("Ingrese el nombre de usuario: ")
    contraseña = input("Ingrese la contraseña: ")
    rol = input("Ingrese el rol (admin/vendedor): ").lower()

    nuevos_usuarios = {
        "ID": id_usuario,
        "Usuario": nombre_usuario,
        "Contraseña": contraseña,
        "Rol": rol
    }
    usuarios.append(nuevos_usuarios)
    print("Usuario registrado exitosamente.")

# Función para ver resumen de ventas
def ver_resumen_ventas():
    if not ventas:
        print("No hay ventas registradas.")
        return

    total_ventas = 0
    print("Resumen de ventas:")
    for venta in ventas:
        print("ID Producto:", venta['IDProducto'], " Cantidad:", venta['Cantidad'], " Vendedor:",  venta['Vendedor'], " Total:", venta['Total'])
       # print(f"ID Producto: {venta['IDProducto']}, Cantidad: {venta['Cantidad']}, Vendedor: {venta['Vendedor']}, Total: ${venta['Total']:.2f}")
        total_ventas += venta['Total']

    print(f"Total generado en ventas: ${total_ventas:.2f}")

# Programa principal del Sistema de Inventario Shaman Tools
while True:
    print("\n--- Sistema de Inventario Shaman Tools ---")
    print("    ---bienvenido, que desea realizar---")
    print("1. Iniciar sesión")
    print("2. Salir")
    opcion = int(input("Seleccione una opción: "))

    if opcion == 1:
        usuario = iniciar_sesion()
        if usuario:
            if usuario["Rol"] == "admin":
                while True:
                    print("\nMenú Administrador")
                    print("1. Registrar a otros usuarios")
                    print("2. Agregar/Actualizar producto")
                    print("3. Consultar inventario")
                    print("4. Ver resumen de ventas")
                    print("5. Salir")
                    opcion_admin = int(input("Seleccione una opción: "))
                    if opcion_admin == 1:
                        registrar_usuario()
                    elif opcion_admin == 2:
                        agregar_actualizar_producto()
                    elif opcion_admin == 3:
                        consultar_inventario()
                    elif opcion_admin == 4:
                        ver_resumen_ventas()
                    elif opcion_admin == 5:
                        break
                    else:
                        print("Opción no válida.")
            elif usuario["Rol"] == "vendedor":
                while True:
                    print("\nMenú Vendedor")
                    print("1. Consultar inventario")
                    print("2. Vender productos")
                    print("3. Salir")
                    opcion_vendedor = int(input("Seleccione una opción: "))
                    if opcion_vendedor == 1:
                        consultar_inventario()
                    elif opcion_vendedor == 2:
                        registrar_venta(usuario)
                    elif opcion_vendedor == 3:
                        break
                    else:
                        print("Opción no válida.")
    elif opcion == 2:
        print("Saliendo del sistema...")
        print("¡Hasta luego!")

        break
    else:
        print("Opción no válida.")
```
